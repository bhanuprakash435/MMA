```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# BGR image

```python
img = cv2.imread('rgb.png')
plt.imshow(img)
```

# BGR to Gray

```python
bgr_img = cv2.imread('rgb.png')
gray_img = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2GRAY)
cv2.imwrite('rgb_to_gray.jpg',gray_img)

plt.imshow(gray_img, cmap = plt.get_cmap('gray'))
```

# BGR to RGB

```python
bgr_rgb = cv2.cvtColor((cv2.imread('rgb.png')), cv2.COLOR_BGR2RGB)
plt.imshow(bgr_rgb)
```

OR

```python
bgr_img = cv2.imread('rgb.png')

b,g,r = cv2.split(bgr_img) 
rgb_img = cv2.merge([r,g,b])     # switch it to rgb
    
plt.imshow(rgb_img)
```

# HSV

```python
img_hsv = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2HSV)

blue_min = np.array([85, 60, 60], np.uint8)
blue_max = np.array([150, 255, 255], np.uint8)
threshold_blue_img = cv2.inRange(img_hsv, blue_min, blue_max)

threshold_blue_img = cv2.cvtColor(threshold_blue_img, cv2.COLOR_GRAY2RGB)
plt.imshow(threshold_blue_img)
```

```python
bgr_im = cv2.imread("grass.png")
rgb_im = cv2.cvtColor(bgr_im, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_im) #original image
```

```python
rgb_green = np.uint8([[[0, 255, 0]]])  # 3d array just because this is what cvtColor expects...
hsv_green = cv2.cvtColor(rgb_green, cv2.COLOR_RGB2HSV)[0, 0, :]
print(hsv_green)
```

```python
thresholding_schematics = cv2.cvtColor(cv2.imread("hsv.png"), cv2.COLOR_BGR2RGB)
plt.imshow(thresholding_schematics)
```

```python
# Convert BGR to HSV
hsv_im = cv2.cvtColor(rgb_im, cv2.COLOR_RGB2HSV)

# define range of hue and intensity
lower_th = hsv_green-np.array([70, 200, 200])
upper_th = hsv_green+np.array([30, 0, 0])

# Threshold the HSV image
mask = cv2.inRange(hsv_im, lower_th, upper_th)
plt.imshow(mask)
```

```python
rgb_res = cv2.bitwise_and(rgb_im, rgb_im, mask=mask)

plt.imshow(rgb_res)
```

# YUV

```python
im = cv2.imread("house.jpg")
rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
yuv = cv2.cvtColor(rgb, cv2.COLOR_BGR2YUV)
y, u, v = cv2.split(yuv)

cv2.imshow('y', y)
cv2.imshow('u', u)
cv2.imshow('v', v)

# Note: Please click ESC to close the output image, else kernel may crash
k = cv2.waitKey(0) & 0xFF
if k == 27:        
    cv2.destroyAllWindows()
```

YUV coding converts the RGB signal to an intensity component (Y) that ranges from black to white plus two other components (U and V) which code the color. 

```python
yuv_img = cv2.imread("yuv.png")
yuv_img = cv2.cvtColor(yuv_img, cv2.COLOR_BGR2RGB)
plt.imshow(yuv_img)
```

# CMY

```python
img_rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
# C = 1 - R
# M = 1 - G
# Y = 1 - B
```

```python
r, g, b = cv2.split(img_rgb)
c = 1 - r / 255
m = 1 - g / 255
y = 1 - b / 255

cv2.imshow('c', c)
cv2.imshow('m', m)
cv2.imshow('y', y)

# Note: Please click ESC to close the output image, else kernel may crash
k = cv2.waitKey(0) & 0xFF
if k == 27:        
    cv2.destroyAllWindows()
```
